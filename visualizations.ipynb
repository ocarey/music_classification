# Visualizations

```python
# Imports
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import pickled Nickelback & Pitbull song dataframe
df = pickle.load(open('df_music.pkl', 'rb'))
```

```python
# Schema
df.info()
```

```python
# Pair plot
sns.pairplot(df, hue = 'Artist')
plt.savefig('Pairplot.png')
```

# Tempo

```python
# Spectral Centroid Mean vs. Artist
dims = (16.0,8.0)
fig, ax = plt.subplots(figsize=dims)
sns.violinplot(x = 'Artist', y = 'Tempo', data = df, ax = ax, color = '#75bbfd', showfliers = False)
plt.rcParams["axes.labelsize"] = 16
plt.rcParams["xtick.labelsize"] = 14
plt.rcParams["ytick.labelsize"] = 14
plt.xlabel('Artist')
plt.ylabel('Tempo')
plt.savefig('Tempo.png')
plt.show()
```

# 3D Plot

```python
# Imports
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale

# Set the dimensions
dims = (10,10)
fig = plt.figure(figsize = dims)
ax = fig.add_subplot(111, projection='3d')

# Scale the features
X = df.drop('Artist', axis = 1)
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
col = list(df.drop('Artist', axis=1).columns)
scaled_X = pd.DataFrame(scaled_X, columns = col)

# Scaled df
y = df['Artist']
scaled_df = pd.concat([y, scaled_X], axis = 1)

# Seperate out Nickelback and Pitbull dataframes
nickelback_df = scaled_df[scaled_df['Artist'] == 'Nickelback']
pitbull_df = scaled_df[scaled_df['Artist'] == 'Pitbull']

# Plot the Nickelback dataframe
ax.scatter(nickelback_df['RMSE Mean'],
           nickelback_df['RMSE Std'],
           nickelback_df['Tempo'],
          color = '#75bbfd',
          label = 'Nickelback')

# Plot the Pitbull dataframe
ax.scatter(pitbull_df['RMSE Mean'],
           pitbull_df['RMSE Std'],
           pitbull_df['Tempo'],
          color = '#ff796c',
          label = 'Pitbull')

# Hide grid lines
ax.grid(False)

# Show the legend
plt.legend()

# Set the axis labels
ax.set_xlabel('RMSE Mean', labelpad = 20)
ax.set_ylabel('RMSE Std', labelpad = 20)
ax.set_zlabel('Tempo Mean', labelpad = 20)

# Extend the layout
plt.tight_layout()
```
